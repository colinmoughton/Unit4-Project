## Baseline Model - Add Early Stopping

````python
from tensorflow.keras.callbacks import EarlyStopping


# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Preprocessing
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# Build baseline model using only dense layers
baseline_model = Sequential([
    Flatten(input_shape=(32, 32, 3)),  # Flatten the 32x32x3 input into a 1D array
    Dense(512, activation='relu'),    # Fully connected layer
    Dense(256, activation='relu'),    # Another fully connected layer
    Dense(128, activation='relu'),    # Another fully connected layer
    Dense(10, activation='softmax')   # Output layer with 10 classes
])

# Compile model
baseline_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Set up early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train model with early stopping
history = baseline_model.fit(
    train_images, 
    train_labels, 
    epochs=30, 
    batch_size=16, 
    validation_split=0.2, 
    callbacks=[early_stopping]  # Include the early stopping callback
)

# Evaluate model
test_loss, test_acc = baseline_model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')
````

````python
plot_accuracy_loss(history, title="Baseline Model- Early Stopping")
````

## Baseline model - add Data Augmentation

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Preprocessing
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# Data augmentation setup
datagen = ImageDataGenerator(
    rotation_range=20,         # Random rotations from -20 to +20 degrees
    width_shift_range=0.2,     # Random horizontal shifts (20% of the image width)
    height_shift_range=0.2,    # Random vertical shifts (20% of the image height)
    shear_range=0.2,           # Random shear transformation
    zoom_range=0.2,            # Random zoom
    horizontal_flip=True,      # Random horizontal flips
    fill_mode='nearest'        # How to fill pixels after a transformation (nearest, constant, etc.)
)

# Fit the data generator on the training images
datagen.fit(train_images)

# Build baseline model using only dense layers
baseline_model = Sequential([
    Flatten(input_shape=(32, 32, 3)),  # Flatten the 32x32x3 input into a 1D array
    Dense(512, activation='relu'),    # Fully connected layer
    Dense(256, activation='relu'),    # Another fully connected layer
    Dense(128, activation='relu'),    # Another fully connected layer
    Dense(10, activation='softmax')   # Output layer with 10 classes
])

# Compile model
baseline_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model with data augmentation
history = baseline_model.fit(
    datagen.flow(train_images, train_labels, batch_size=16),  # Using data augmentation generator
    epochs=30, 
    validation_data=(test_images, test_labels)  # Using test data for validation (no augmentation)
)

# Evaluate model
test_loss, test_acc = baseline_model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')
````

````python
plot_accuracy_loss(history, title="Baseline Model- Data Augmentation")
````

## Explanation of Parameters:

1. Learning Rate: You can set a custom learning rate (e.g., 0.001 or 0.0005), which controls how quickly the model learns. A smaller learning rate can result in slower training, but may lead to more stable convergence, while a larger learning rate can speed up training but risks overshooting the optimal solution.

2. Beta 1 and Beta 2: These parameters control the moving averages of the gradient (first moment) and the squared gradient (second moment).
       - Beta 1 (0.9): Controls how much of the previous gradients are considered in the current update.
       - Beta 2 (0.999): Controls how much of the previous squared gradients are considered. Both are typically left at their default values (0.9 for beta_1 and 0.999 for beta_2), but you can adjust them if needed.

3. Epsilon: This is a small constant added to prevent division by zero in some calculations. You typically don't need to change this value unless you're dealing with numerical stability issues.

4. Decay: This is the learning rate decay. If you want to reduce the learning rate gradually during training (e.g., after each epoch), you can set a value for decay (e.g., 0.0001). This helps to refine the model's performance after training for a number of epochs.

## Example with Learning Rate Decay:

````python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers.schedules import ExponentialDecay


# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Preprocessing
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# Build baseline model using only dense layers
baseline_model = Sequential([
    Flatten(input_shape=(32, 32, 3)),  # Flatten the 32x32x3 input into a 1D array
    Dense(512, activation='relu'),    # Fully connected layer
    Dense(256, activation='relu'),    # Another fully connected layer
    Dense(128, activation='relu'),    # Another fully connected layer
    Dense(10, activation='softmax')   # Output layer with 10 classes
])

# Define an exponential learning rate decay
learning_rate_schedule = ExponentialDecay(
    initial_learning_rate=0.001,  # Starting learning rate
    decay_steps=100000,            # Number of steps before decay
    decay_rate=0.96,               # The rate at which the learning rate decays
    staircase=True                 # Whether the decay should happen in discrete intervals
)

# Compile the model with the learning rate schedule
optimizer = Adam(learning_rate=learning_rate_schedule)

model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(train_images, train_labels, epochs=30, batch_size=16, validation_split=0.2)

# Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')
````

## Report: Understanding the Challenges and Alternatives for Image Classification Using Dense Neural Networks
### 1. Why Cant We Reach Good Accuracy on This Task?

The difficulty in achieving high accuracy on image classification tasks with a fully connected (dense) neural network can be attributed to the following factors:

- Loss of Spatial Information: Fully connected layers treat each pixel in the image as independent, ignoring the spatial relationships between neighboring pixels. This leads to a loss of critical information about patterns, edges, and textures that are essential for identifying objects in images.

- High Dimensionality: CIFAR-10 images have dimensions 3232332323 (3072 features). A fully connected network must learn relationships between all these features, which increases the complexity and computational cost, making it harder to optimize the model.

- Overfitting: Dense networks have a large number of parameters, especially for high-dimensional data. This increases the risk of overfitting, particularly when the dataset is relatively small (like CIFAR-10).

- Activation Function Saturation: Using activation functions such as sigmoid can lead to saturation issues, where gradients become very small during backpropagation. This slows down learning and reduces the network's ability to model complex relationships.

### 2. Why Are Fully Connected Neural Networks Not Efficient for Image Tasks?

Fully connected neural networks are inefficient for image tasks for several reasons:

- Flat Representation: Flattening images into a 1D vector removes the hierarchical and spatial structure of the data. This prevents the network from learning local patterns, such as edges, corners, and textures.

- Parameter Explosion: A fully connected layer requires a weight for every connection between neurons. For an image of size 3232332323, the first dense layer alone would have 3072n3072n weights (where nn is the number of neurons), leading to excessive memory usage and training time.

- Lack of Translation Invariance: Dense layers lack the ability to recognize patterns regardless of their position in the image. This means the network must learn redundant features for every possible position, making it inefficient.

### 3. What Architecture Can Be Used for Such Tasks? Why Are They More Adapted?

A Convolutional Neural Network (CNN) is a much better-suited architecture for image classification tasks.

Why CNNs Are More Adapted:

- Spatial Hierarchy and Locality: CNNs use convolutional layers that apply filters to local regions of the image. This allows the network to learn spatial hierarchies, such as edges in the first layer and complex shapes in deeper layers.

- Parameter Sharing: The same filter is applied across the entire image, drastically reducing the number of parameters compared to fully connected layers. This makes CNNs more efficient and less prone to overfitting.

- Translation Invariance: CNNs can detect patterns (like edges or textures) anywhere in the image, providing robustness to changes in object position.

- Pooling Layers: Pooling layers in CNNs reduce the spatial dimensions of the feature maps, helping to retain the most important information while reducing computational complexity.

Popular Architectures for Image Tasks:

- LeNet-5: Early CNN architecture, effective for small datasets.
- AlexNet: Introduced deeper architectures and made use of GPUs for large-scale datasets.
- VGGNet: Focuses on very deep networks with small convolutional filters.
- ResNet: Uses residual connections to train very deep networks without vanishing gradients.

Comparison With Dense Networks: While dense networks lack the ability to leverage spatial structure, CNNs are designed specifically to handle image data. They significantly outperform dense networks in tasks like CIFAR-10 classification, achieving accuracies well above 90% with appropriate tuning.
***
## Conclusion

Fully connected networks are inherently limited for image classification tasks due to their inability to capture spatial relationships, inefficient parameter usage, and susceptibility to overfitting. Convolutional Neural Networks overcome these limitations by leveraging spatial hierarchies, parameter sharing, and translation invariance, making them the gold standard for image-related tasks like CIFAR-10 classification. Transitioning to a CNN architecture is essential for achieving high accuracy and efficient learning on image data.

## Plot function

````python
import matplotlib.pyplot as plt

def plot_accuracy_loss(history, title="Model Performance"):
    """
    Plots training and validation accuracy and loss from the history object.

    Parameters:
    - history: Keras History object from model.fit()
    - title: Title of the plot
    """
    # Extract accuracy and loss
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    # Create figure
    epochs = range(1, len(acc) + 1)
    plt.figure(figsize=(12, 6))
    
    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, label='Training Accuracy', marker='o')
    plt.plot(epochs, val_acc, label='Validation Accuracy', marker='o')
    plt.title(f"{title} - Accuracy")
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, label='Training Loss', marker='o')
    plt.plot(epochs, val_loss, label='Validation Loss', marker='o')
    plt.title(f"{title} - Loss")
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    # Show plot
    plt.tight_layout()
    plt.show()

# Example usage:
# Assuming `history` is the output from model.fit()
# plot_accuracy_loss(history, title="Baseline Model")
````
